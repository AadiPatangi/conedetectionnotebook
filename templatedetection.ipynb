```python

 # Read the main image
img_rgb = cv.imread(img_path)
 
# Convert it to grayscale
img_gray = cv.cvtColor(img_rgb, cv.COLOR_BGR2GRAY)
 
# Read the template
template = cv.imread(tem_path, 0)
 
# Store width and height of template in w and h
w, h = template.shape[::-1]
 
# Perform match operations.
res = cv.matchTemplate(img_gray, template, cv.TM_CCOEFF_NORMED)
 
# Specify a threshold
threshold = 0.75
# Store the coordinates of matched area in a numpy array
loc = np.where(res >= threshold)
 
# Draw a rectangle around the matched region.
for pt in zip(*loc[::-1]):
    cv.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0, 255, 255), 1)
    
    w1 = int((2*pt[0] + w )/2) - 10
    w2 = int((2*pt[0] + w )/2) + 10
    h1 = int((2*pt[1] + h )/2) - 10
    h2 = int((2*pt[1] + h )/2) - 10

    print(w1,w2,h1,h2)
    print(img_rgb[w1,h1,2],img_rgb[w1,h1,0])
    avg_red = np.average(img_rgb[w1:w2,h1:h2,2])
    avg_blue = np.average(img_rgb[w1:w2,h1:h2,0])
    #avg_color_per_row = np.average(img_rgb[w1:w2,h1:h2], axis=0)
    #avg_color = np.average(avg_color_per_row, axis=0)

    #print(avg_color)
 
# Show the final image with the matched area.
#cv2.imshow('Detected', img_rgb)
```
